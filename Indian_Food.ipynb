# INDIAN FOOD DATASET

```python
#Imported all the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the csv file
ind = pd.read_csv("C:\\Users\\HP\\Desktop\\TERM-3\\DV\\Group5-indian_food.csv", na_values=[-1])
```

```python
#To display maximum rows and columns we have used the set_option
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
ind
```

```python
#Dictionary named values has been created which will replace -1 values 
#in the prep_time and cook_time columns with their respective mean values
#and it will replace -1 values in the categorical with the item occuring maximum
#number of times
values = {'prep_time': np.mean(ind['prep_time']), 
          'cook_time':np.mean(ind['prep_time']), 
          'flavor_profile':ind['flavor_profile'].value_counts().index[0], 
          'course' :ind['course'].value_counts().index[0], 
          'state' :ind['state'].value_counts().index[0], 
          'region':ind['region'].value_counts().index[0]}
```

```python
#A new variable has been created which will use the dictionary defined above
new_ind = ind.fillna(value=values)
```

# CLEANED DATASET

```python
new_ind.head()
```

# BAR GRAPH shows the relationship between cooking time and state

```python
plt.figure(figsize = (30,16))
ax= sns.barplot(x="state",y="cook_time",data=new_ind)
ax.set_title("Bar Plot - Time Taken by different states to prepare food")
plt.show()

#INTERPRETATION
#The maximum time taken to prepare food was for Bihar
#The minimum time taken to prepare food was for Nagaland
```

# BARPLOT shows the division between Non Veg and Veg Prferences across Different Regions

```python
br=sns.barplot(x="region", y="prep_time", hue="diet", data=new_ind)
br.set_title("Bar Plot - To show the division between Non Veg and Veg Prferences across Different Regions ")
plt.show()
#INTERPRETATION
# Here Barplot is used to show the relationship between the numeric and categorical variable 
#i.e. the number of Vegetarians and Non-Vegetarians across different Regions.
# The wick on top of the rectangles depicts indication of the uncertainty
#pertaining to the reliability of the Data.
```

# HEATMAP depicts the relationship between the flavor profile and type of diet

```python
tab = pd.crosstab(new_ind['diet'],new_ind['flavor_profile'])
print(tab)
htmp=sns.heatmap(tab, cmap="rocket")
htmp.set_title("Heat Map - To show the division between Non Veg and Veg Prferences across Different Flavour Profiles ")
plt.show()
#Interpretation
#There is no relation between bitter and vegetarian diet as well as non vegetarian diet 
#There exists a significant relationship between sweetness in food and in vegetarian diet
#There exists a moderate relationship between spicy food and non vegetarian diet
```

# PAIRPLOT shows the relationship between preparation time and cook time grouped by course

```python
sns.set_style("ticks")
sns.pairplot(new_ind,hue='course',diag_kind="kde",kind="scatter",palette="husl")
plt.show()
#INTERPRETATION
#A matrix is produced that shows relationship between each variable in our data
#that is., cooking time and preparation time which has been grouped by course 
#course includes dessert, main course, starter, snack 
```

# RELATIONALPLOT shows the relationship between region and preparation time grouped by course

```python
#Relplot is made between region and preparation time where the data has been grouped on the basis of course
#On the x axis region is taken and on the y axis preparation time
sns.relplot(x="region",y="prep_time",hue='course',size_order=[100,100],s=100,data=new_ind)
#We can see that dessert takes the maximum time to prepare
#in the east and main course takes the maximum time 
#to prepare in the south and north
```

# DATASET CONTAINING FLAVOR PROFILE REGION WISE

```python
#Groupby function is used to group the data on the basis of region and flavor profile
#s is assigned to the grouoped data where the time taken to prepare food as per 
#the flavour profile has been summated
#df is a new data frame that is created to store the value of s
# fillna is used to fill the empty values with 0
new_ind_g= new_ind.groupby(['region','flavor_profile'])
s= new_ind_g['prep_time'].sum().unstack()
df = pd.DataFrame(s)
df.fillna(value=0)
```

```python
#Index are reset with this function so as to use them as labels in the pie chart
dv= df.reset_index("region")
```

```python
#The data of the sweet food category is converted into percentage
dv['percent_sweet'] = (dv['sweet'] / dv['sweet'].sum()) * 100
```

# PIECHART depicting number of sweet dishes in the various regions

```python
#The explode function is used to explode the east and the centeral region
#The explode function is used to explode the east and the centeral region
#Pie chart is made for the depiction of the number of sweet dishes that are made in each region
explode= [0.2,0.1,0,0,0,0]
plt.pie(dv['sweet'],labels=dv['region'],autopct='%.2f',explode=explode)
plt.show()
#INTERPRETATION
#The pie chart shows that the maximum sweet dishes are from in East and the Least are from central
```

# JOINTPLOT depicts the relationship between preparation time and cooking time 

```python
#Seaborns jointplot displays a relationship between 2 variables (bivariate) 
#as well as 1D profiles (univariate) in the margins. 
sns.jointplot(x='prep_time',y='cook_time',data=new_ind,kind='kde')
plt.show()
```

# FACETGRID shows cooking time as per course(i.e.,dessert, main course, starter, snack)

```python
#FacetGrid class helps in visualizing distribution of one variable as well as 
#the relationship between multiple variables separately within subsets of your 
#dataset using multiple panels.
b=sns.FacetGrid(new_ind,col="course")
b.map(plt.hist,"cook_time")
```

# STATIC AND INTERACTIVE PLOT MADE USING BOKEH

```python
#Imported Bokeh for plotting Static and Interactive Plot 
from bokeh.io import curdoc, output_notebook
from bokeh.plotting import figure, show
from bokeh.models import HoverTool, ColumnDataSource, CategoricalColorMapper, Slider
from bokeh.palettes import Spectral6
from bokeh.layouts import widgetbox, row
```

```python
#Loaded the output notebook to display the Static and Interactive Plot
output_notebook()
```

```python
#this is used to create a list of the region we want to plot
regions_list = new_ind.region.unique().tolist()
```

```python
#this is used to allot different colours to the various regions in the list
color_mapper = CategoricalColorMapper(factors=regions_list, palette=Spectral6)
```

```python
#data source is made to plot the data
#ColumnDataSource function is used to 
source = ColumnDataSource(data={
    'x': new_ind['prep_time'],
    'y': new_ind['cook_time'],
    'state': new_ind['state'],
    'region': new_ind['region']})

#source = ColumnDataSource(data={
#    'x': new_ind['prep_time'],
 #   'y': new_ind['cook_time'],
  #  'state': new_ind['state'],
   # 'region': new_ind['region']})
```

```python
#The minmum and maximum value of preparation time are saved in xmin and xmax
xmin, xmax = min(new_ind['prep_time']), max(new_ind['prep_time'])
#The minmum and maximum value of cooking time are saved in ymin and ymax
ymin, ymax = min(new_ind['cook_time']), max(new_ind['cook_time'])
```

```python
#This is used to create a figure:plot
#plot height and width are provided
#x range and y range are taken to avoid random values 
plot = figure(title='x write here', 
              plot_height=600, plot_width=1000,
              x_range=(xmin, xmax),
              y_range=(ymin, ymax), y_axis_type='log')
```

```python
#This is used to add circles on the plot, fill_alpha gives the transparency
plot.circle(x='x', y='y', fill_alpha=0.8, source=source, legend_label='region',
            color=dict(field='region', transform=color_mapper),
            size=7)
```

```python
show(plot) #This gives the static figure
```

```python
# Adding Slider with minimum and maximum preparation time 

slider = Slider(start=min(new_ind['prep_time']), end=max(new_ind['prep_time']), step=1, value=min(new_ind['prep_time']), title='Preparation Time')


def update_plot(attr, old, new):
    # set the `yr` name to `slider.value` and `source.data = new_data`
    pt = slider.value

    new_data = data={
    'x': new_ind['prep_time'],
    'y': new_ind['cook_time'],
    'state': new_ind['state'],
    'region': new_ind['region']}
    
    
    
    
 #   {
  #      'x': data.gdp[data['year'] == yr],
   #     'y': data.co2[data['year'] == yr],
    #    'country': data.country[data['year'] == yr],
     #   'region': data.region[data['year'] == yr],
    #}
    source.data = new_data
```

```python
# Adding title to figure: plot.title.text
plot.title.text = 'x type here %d' 
# % yr
```

```python
# Attach the callback to the 'value' property of slider
slider.on_change('value', update_plot)

layout = row(widgetbox(slider), plot)
curdoc().add_root(layout)
```

```python
# Adding the Hover tool

# Create a HoverTool: hover
hover = HoverTool(tooltips=[('Name', '@name'), ('State', '@state'), ('region', '@region')])

# Add the HoverTool to the plot
plot.add_tools(hover)
```
